```python
class Celeb(Dataset):
  def __init__(self,dir,transform=None):
    super().__init__()
    self.dir=dir
    self.transform=transform
  def __len__(self):
    return len(os.listdir(self.dir))
  def __getitem__(self,idx):
    x=os.listdir(self.dir)[idx]
    img=read_image(os.path.join(self.dir,x))
    if self.transform:
      return self.transform(img)
    else:
      return img
```

```python
dir='/content/img_align_celeba/img_align_celeba'
trans=v2.Compose([v2.ToDtype(torch.float32,scale=True),v2.Resize((64,64),antialias=True)])
data=Celeb(dir,trans)
```

```python
epochs=20
batch=64
lr=1e-4
beta1=0.5
beta2=0.999
lamb=10
n_critic=5
dim,n_channels,img_size=100,3,64
dataloader=DataLoader(data,shuffle=True,batch_size=batch,drop_last=True)
```

```python
x=next(iter(dataloader))
grid=make_grid(x)
plt.figure(figsize=(10,5))
plt.imshow(grid.permute(1,2,0).clamp(0.,1.))
plt.axis('off')
plt.show()
```

```python
class Generator(nn.Module):
    def __init__(self,dim,n_channels,img_size):
        super().__init__()
        self.dim=dim
        self.net=nn.Sequential(
            nn.ConvTranspose2d(dim,img_size*4,4,1,0,bias=False),
            nn.BatchNorm2d(img_size*4),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(img_size*4,img_size*2,4,2,1,bias=False),
            nn.BatchNorm2d(img_size*2),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(img_size*2,img_size*2,4,2,1,bias=False),
            nn.BatchNorm2d(img_size*2),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(img_size*2,img_size,4,2,1,bias=False),
            nn.BatchNorm2d(img_size),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(img_size,n_channels,4,2,1,bias=False),
            nn.Tanh(),
        )
    def forward(self,x):
        return self.net(x)
```

```python
class Critic(nn.Module):
    def __init__(self,n_channels,img_size):
        super().__init__()
        self.net=nn.Sequential(
            nn.Conv2d(n_channels,img_size,4,2,1,bias=False),
            nn.LayerNorm((img_size,img_size//2,img_size//2)),
            nn.LeakyReLU(),
            nn.Conv2d(img_size,img_size*2,4,2,1,bias=False),
            nn.LayerNorm((img_size*2,img_size//4,img_size//4)),
            nn.LeakyReLU(),
            nn.Conv2d(img_size*2,img_size*2,4,2,1,bias=False),
            nn.LayerNorm((img_size*2,img_size//8,img_size//8)),
            nn.LeakyReLU(),
            nn.Conv2d(img_size*2,img_size*4,4,2,1,bias=False),
            nn.LayerNorm((img_size*4,img_size//16,img_size//16)),
            nn.LeakyReLU(),
            nn.Conv2d(img_size*4,1,4,1,0,bias=False),
        )
    def forward(self,x):
        return self.net(x)
```

```python
def weight_init(mod):
    name=mod.__class__.__name__
    if 'Conv' in name:
        nn.init.normal_(mod.weight.data,0.,0.02)
    elif 'BatchNorm' in name:
        nn.init.normal_(mod.weight.data,1.,0.02)
        nn.init.constant_(mod.bias.data,0.)
```

```python
device='cuda' if torch.cuda.is_available() else 'cpu'
```

```python
gen=Generator(dim,n_channels,img_size).to(device)
critic=Critic(n_channels,img_size).to(device)
gen.apply(weight_init)
critic.apply(weight_init)
summary(gen,input_size=(100,1,1)),summary(critic,input_size=(3,64,64))
```

```python
c_optim=optim.Adam(critic.parameters(),lr=lr,betas=(beta1,beta2))
g_optim=optim.Adam(gen.parameters(),lr=lr,betas=(beta1,beta2))
```

```python
def gradient_penalty(real,fake,critic,device='cpu'):
  batch=real.shape[0]
  epsilon=torch.rand(batch,1,1,1).to(device)
  interpolated_imgs=epsilon*real+(1-epsilon)*fake
  interpolated_imgs.requires_grad_(True)
  i_pred=critic(interpolated_imgs).reshape(-1)
  gradients=torch.autograd.grad(inputs=interpolated_imgs,
                                outputs=i_pred,
                                grad_outputs=torch.ones_like(i_pred),
                                create_graph=True,
                                retain_graph=True)[0]
  grads=gradients.reshape(batch,-1)
  grads_norm=grads.norm(2,dim=1)
  g_penalty=grads_norm.sub(1).pow(2).mean()
  return g_penalty
```

```python
print('Training......')
new_imgs=[]
c_l=[]
g_l=[]
for epoch in range(epochs):
  gen.train()
  critic.train()
  for idx,imgs in enumerate(dataloader):
    input=imgs.to(device)
    for c in range(n_critic):
      critic.zero_grad()
      noise=torch.randn(batch,dim,1,1,device=device)
      fake=gen(noise)
      loss_real=critic(input).reshape(-1)
      loss_fake=critic(fake.detach()).reshape(-1)
      loss_c=loss_fake.mean()-loss_real.mean()+lamb*gradient_penalty(input,fake.detach(),critic,device=device)
      loss_c.backward()
      c_optim.step()
    gen.zero_grad()
    g_pred=critic(fake).reshape(-1)
    loss_gen=-(g_pred.mean())
    loss_gen.backward()
    g_optim.step()
    c_l.append(loss_c.item())
    g_l.append(loss_gen.item())
    if idx%1000==0:
      print(f'batch:{idx},  critic_loss:{loss_c.item()} ,  gen_loss:{loss_gen.item()}')
  print(f'***epoch:{epoch},  critic_loss:{loss_c.item()} ,  gen_loss:{loss_gen.item()}***')
  if (epoch+1)%10==0:
    with torch.no_grad():
      n_noise=torch.randn(64,100,1,1,device=device)
      n_imgs=gen(n_noise)
      new_imgs.append(make_grid(n_imgs.detach().cpu(),normalize=True))
```
